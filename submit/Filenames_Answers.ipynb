```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_pickle('C:/Users/Boonchong/Desktop/Intern/iecoe-sg-technical-interview/data/df_cases_200906.gzip')
df_label = pd.read_pickle('C:/Users/Boonchong/Desktop/Intern/iecoe-sg-technical-interview/data/df_label_200906.gzip')
```

```python
df.head()
```

```python
df_label.head()
```

```python
# Making new columns to concantenate the FileNames
df['InvalidFileNames'] = ""
df['ValidFileNames'] = ""
```

```python
# Dropping columns that are not used
df = df.drop(['Language','StartDate','DocumentType'],axis=1)
```

```python
df
```

```python
# Finding InvalidFileNames
df.loc[(df.QualityScore < 0.81) | (df.IsExecuted == False ), 'InvalidFileNames'] = df['FileName']

# ~(A U C) = (~A) n (~C)
df.loc[(df.QualityScore >= 0.81) & (df.IsExecuted == True ), 'ValidFileNames'] = df['FileName']
```

```python
# Using this as the base case
df.loc[df['CaseId'] == '003061230728']
```

```python
# Removing OcrText for FileNames that are not valid
df.loc[(df.ValidFileNames == ""),'OcrText'] = ""
```

```python
# Check if the base case is updated
df.loc[df['CaseId'] == '003061230728']
```

```python
# The 2 columns are no longer needed
df = df.drop(['IsExecuted','QualityScore'],axis=1)
```

```python
# Check updated base case
df.loc[df['CaseId'] == '003061230728']
```

```python
# Joining all InvalidFileNames and ValidFileNames together respectively as well as OcrText

df.InvalidFileNames = df.groupby('CaseId')['InvalidFileNames'].transform(lambda x : ', '.join(x))
df.ValidFileNames = df.groupby('CaseId')['ValidFileNames'].transform(lambda x : ', '.join(x))

df.OcrText = df.groupby('CaseId')['OcrText'].transform(lambda x : ''.join(x))
```

```python
# Making sure the base case is updated
df.loc[df['CaseId'] == '003061230728']
```

```python
# Looking at InvalidFileNames after concantenating, there are extra ', '
df.iloc[917,3]
```

```python
# Checking for duplicates
df.head()
```

```python
# Remove FileName, so now dataset has duplicates
df = df.drop(['FileName'],axis=1)
```

```python
# Dropping duplicates
df = df.drop_duplicates()
```

```python
# Check if the duplicates are dropped
df.head()
```

```python
# Adding '[]' to the prefix and suffix of our FileNames as well as replacing ', ' with '[]' 

df['InvalidFileNames'] = '[' + df['InvalidFileNames'].astype(str) + ']'
df['ValidFileNames'] = '[' + df['ValidFileNames'].astype(str) + ']'
df.loc[(df.InvalidFileNames == '[, ]'), 'InvalidFileNames'] = '[]'
df.loc[(df.ValidFileNames == '[, ]'), 'ValidFileNames'] = '[]'
```

```python
# Check if FileNames are replaced and updated
df.head()
```

```python
# Base Case
df.loc[df['CaseId'] == '003061230728'].iloc[0,2]
```

```python
# Since we have '[xxxxx, , xxxxx]'
# Expected outcome is '[xxxxx, xxxxx]'
# Therefore ' ,' is needed to be replaced


df.InvalidFileNames = df.InvalidFileNames.str.replace(" ,", "")
df.ValidFileNames = df.ValidFileNames.str.replace("\[, ", "[")
```

```python
# Checking with base case
df.loc[df['CaseId'] == '003061230728'].iloc[0,2]
```

```python
df.head()
```

```python
# Merging with labels 
result = pd.merge(df, df_label, on=['CaseId'])
```

```python
result.head()
```

```python
# Rearranging Columns
result = result[['CaseId', 'InvalidFileNames', 'ValidFileNames', 'OcrText', 'label_1', 'label_2']]
```

```python
result.head()
```

```python
result = result.sort_values(by='CaseId',ignore_index=True)
```

```python
# Checking with example
result.iloc[1093:1098,:]
```

```python
# Exporting as df_final.gzip
result.to_pickle('df_final.gzip')
```
